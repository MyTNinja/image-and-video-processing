```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from skimage.color import rgb2gray
from skimage import data
import numpy as np
```

```python
image = data.brick()
```

```python
r1, s1 = 100, 30
r2, s2 = 200, 220

alpha = s1/r1
beta = (s2 - s1)/(r2 - r1)
gamma = (255 - s2)/(255 - r2)

image_cs = image.copy()

for i in range(image.shape[0]):
  for j in range(image.shape[1]):
    r = image[i][j]

    if 0 <= r and r <= r1:
      image_cs[i][j] = alpha*(r)
    elif r1 < r and r <= r2:
      image_cs[i][j] = beta*(r - r1) + s1
    else:
      image_cs[i][j] = beta*(r - r2) + s2
```

```python
image_thresh = image_cs.copy()
thresh = 150

for i in range(image_cs.shape[0]):
  for j in range(image_cs.shape[1]):
    r = image_cs[i][j]
    if r > thresh:
      image_thresh[i][j] = 255
    else:
      image_thresh[i][j] = 0
```

```python
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.title('Original Image')
plt.axis('off')
plt.imshow(image, cmap = 'gray')

plt.subplot(2, 2, 3)
plt.title('Contrast Stretched')
plt.axis('off')
plt.imshow(image_cs, cmap = 'gray')

plt.subplot(2, 2, 4)
plt.title('Binary Image')
plt.axis('off')
plt.imshow(image_thresh, cmap = 'gray')

plt.show()
```

```python
'''
Conclusion:

Image of brick is already a gray image. After contrast stretching it can be noted that
the new image has more darker "bricks" as compared to the original, improving
contrast. The image is further binarised into a black and white image via
thresholding.
'''
```
