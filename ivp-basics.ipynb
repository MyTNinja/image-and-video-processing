```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
img1 = mpimg.imread('shark-1.jpg')

print(img1, img1.shape, img1.dtype, type(img1))
```

```python
img2 = mpimg.imread('shark-2.jpg')

print(img2, img2.shape, img2.dtype, type(img2))
```

```python
plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(img1)
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(img2)
plt.axis('off')
plt.show()
```

```python
img2_modify = img2.copy()
img2_modify[img2_modify<100]=0

plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(img2)
plt.title('Original')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.title('Darker')
plt.imshow(img2_modify)
plt.axis('off')
plt.show()
```
