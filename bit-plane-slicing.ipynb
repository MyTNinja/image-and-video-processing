```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
img = mpimg.imread('lake.tif')
img.dtype, img.shape, type(img)
```

```python
imgbit = img.copy()
```

```python
def decimalToBinary(n):
    return '{:08b}'.format(int(n))
```

```python
def plot(im, i, title):
  plt.subplot(3, 3, i)
  plt.imshow(im, cmap = 'gray')
  plt.title(title)
  plt.axis('off')
```

```python
bit = img.copy()

plt.figure(figsize=(10, 10))

plot(img, 1, 'Original')

for k in range(8):
  for i in range(imgbit.shape[0]):
    for j in range(imgbit.shape[1]):
      r = list(decimalToBinary(imgbit[i, j]))
      bit[i, j] = int(r[k])
  plot(bit, abs(k-7)+2, 'Bit Plane ' + str(abs(k-7)))
plt.show()
```
